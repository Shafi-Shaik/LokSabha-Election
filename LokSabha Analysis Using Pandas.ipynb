```python
##### Standardizing column names to lower case
LS.columns=LS.columns.str.lower()
LS.columns
```

```python
#### Total Seats or Total Winners
LS["winner"].value_counts()
```

```python
### Average Age of candidates Party-Wise
#SADM, JKN, VPI has the candidates with average age >=70
LS[['party', 'age']].groupby('party').mean('age').sort_values('age', ascending=0) #mean() excludes all NA values, so no need of na.rm
```

<h3> <i> No. of Seats won by each party</i> </h3>

```python
LS[["party", "winner"]].groupby('party').agg ({'winner':[('seats_won','sum'),('seats_contested', 'count')]})
```

<h3> <i> No. of Seats won by State </i> </h3>

```python
LS[['state', 'winner']].groupby('state').agg({'winner':[('Total Seats','sum')]})
```
